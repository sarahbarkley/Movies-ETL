```python
def clean_movie(movie):
    movie = dict(movie) #create a non-destructive copy
    alt_titles = {}
    keys = ['Also known as', 'Arabic', 'Cantonese', 'Chinese', 'French', 'Hangul', 'Hebrew', 'Hepburn', 'Japanese', 
            'Literally', 'Mandarin', 'McCune-Reischauer', 'Original title', 'Polish', 'Revised Romanization', 'Romanized',
            'Russian', 'Simplified', 'Traditional', 'Yiddish']
    for key in keys:
        if key in movie:
            alt_titles[key] = movie[key]
            movie.pop(key)
    if len(alt_titles) > 0:
        movie['alt_titles'] = alt_titles
    return movie
```
